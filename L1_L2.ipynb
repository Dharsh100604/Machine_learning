```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
X_b=np.c_[np.ones((X.shape[0],1)),X]
X_b
```

```python
alpha_ridge=1.0
alpha_lasso=1.0
num_iterations=1000
learning_rate=0.01
```

```python
def lasso_regression(X,y,alpha,num_iterations,learning_rate):#square of coeff.It can give 0 error
  m,n=X.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred=X.dot(theta)
    gradient=(X.T.dot(y_pred-y)+alpha* np.sign(theta))/m
    theta -= learning_rate*gradient
  return theta
```

```python
def ridge_regression(X,y,alpha,num_iterations,learning_rate):#square of coeff
  m,n=X.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred=X.dot(theta)
    gradient=(X.T.dot(y_pred-y)+alpha* (theta))/m
    theta -= learning_rate*gradient
  return theta
```

```python
theta_ridge=ridge_regression(X_b,y,alpha_ridge,num_iterations,learning_rate)
theta_lasso=lasso_regression(X_b,y,alpha_lasso,num_iterations,learning_rate)
```

```python
plt.scatter(X,y,label='Data')
plt.plot(X,X_b.dot(theta_ridge),label=f'L2 Regression',color="red")#ridge
plt.plot(X,X_b.dot(theta_lasso),label=f'L1 Regression',color="blue")#lasso
plt.xlabel(X)
plt.ylabel(y)
plt.legend()
plt.show()
```
