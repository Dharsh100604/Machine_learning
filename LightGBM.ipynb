```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( x, y, test_size=0.25, random_state=28)
```

```python
import lightgbm as lg
```

```python
model=lg.LGBMClassifier()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred) * 100
print("Accuracy  ",accuracy)
```

```python
model_lgb = lg.LGBMClassifier()
from sklearn.model_selection import GridSearchCV
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'bagging_fraction':[0.1,0.2,0.3],
    'min_data_in_leaf': [8, 39, 10],
}
grid_search = GridSearchCV(estimator=model_lgb, param_grid=param_grid, scoring='accuracy', cv=5, verbose=1)
```

```python
from sklearn.model_selection import GridSearchCV
import warnings

# Set up GridSearchCV with error_score='raise'
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, error_score='raise')

# Suppress warnings if needed
warnings.filterwarnings("ignore")

# Fit the grid search
grid_search.fit(X_train, Y_train)

# Print results
print("Best Parameters: ", grid_search.best_params_)
print("Accuracy: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy on Test Set: ", accuracy)
```
