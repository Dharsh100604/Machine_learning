**Identify outliers:**
* Z-score (standard score)
   * mean/std
* IQR(Inter Quartile range)

```python
from scipy import stats
hp_zscore= np.abs(stats.zscore(df["normalized-losses"]))
hp_zscore
```

```python
df['normalized-losses']
```

```python
#set the threshold
threshold=hp_zscore.max()
#position of the outlier
index_val= np.where(hp_zscore >= threshold)
df.iloc[index_val]["normalized-losses"]
```

```python
new_data = df.drop(index=df.iloc[index_val]['normalized-losses'].index) #drop the index that has outlier
new_data
```

```python
#cal upper and lower limits
Q1 = df['length'].quantile(0.25)
Q3 = df['length'].quantile(0.75)
IQR=Q3-Q1
lower=Q1+1.5*IQR
upper=Q3+1.5*IQR
```
