```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
X= data.iloc[:,1:2].values
X
```

```python
Y=data.iloc[:,2].values
Y
```

```python
from sklearn .linear_model import LinearRegression #summa
lin_reg= LinearRegression()
lin_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)#raise degree
X_poly2=poly_reg.fit_transform(X)
lin_reg_2 =LinearRegression()#Then linear regression
lin_reg_2.fit(X_poly2,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=3)#raise degree
X_poly3=poly_reg.fit_transform(X)
lin_reg_3 =LinearRegression()#Then linear regression
lin_reg_3.fit(X_poly3,Y)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title("Linear Regression")
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,Y,color='r')
plt.plot(X,lin_reg_2.predict(X_poly2),color='blue')
plt.plot(X,lin_reg_3.predict(X_poly3),color='hotpink')
plt.title("Linear Regression")
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X=data.drop('Y house price of unit area',axis=1)
X
```

```python
Y=data['Y house price of unit area']
Y
```

```python
poly_reg=PolynomialFeatures(degree=2,include_bias=False)# we no need error(bias)
```

```python
poly_feat=poly_reg.fit_transform(X)
poly_feat.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(poly_feat,Y,test_size=0.3,random_state=101)
```

```python
polynomial=LinearRegression()
```

```python
polynomial.fit(X_train,Y_train)
```

```python
Y_pred=polynomial.predict(X_test)
```

```python
pd.DataFrame({"Y_test": Y_test,"Y_pred": Y_pred,"Residuals": Y_test-Y_pred})
```

```python
poly_MAE=metrics.mean_absolute_error(Y_test,Y_pred)
poly_MSE=metrics.mean_squared_error(Y_test,Y_pred)
poly_RMSE=np.sqrt(poly_MSE)
```
